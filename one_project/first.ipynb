```python
import pandas as pd

file_path = '../test.xlsx'

go_data = pd.read_excel(file_path, sheet_name='출근')
leave_data = pd.read_excel(file_path, sheet_name='퇴근')
```

```python
go_sort_data = go_data.sort_values(by=['탑승 일시', '탑승 시간'], ascending=[True, True])
```

```python
go_sort_data.head()
```

```python
leave_data = leave_data.sort_values(by=['탑승 일시', '탑승 시간'], ascending=[True, True])
leave_data.head()
```

```python
leave_data = leave_data.reset_index(drop=True)
leave_data.index += 1
leave_data.head()
```

```python
go_sort_data = go_sort_data.reset_index(drop=True)
go_sort_data.index += 1
go_sort_data.head()
```

```python
go_sort_data = go_sort_data.drop("차량번호", axis=1)
leave_data = leave_data.drop("차량번호", axis=1)

go_sort_data.head()
```

```python
from datetime import datetime

go_sort_data['탑승 시간'] = pd.to_datetime(go_sort_data['탑승 시간'])
go_sort_data['하차 시간'] = pd.to_datetime(go_sort_data['하차 시간'])

# 하차 시간이 "0시"인 경우 다음 날의 자정으로 처리
go_sort_data.loc[go_sort_data["하차 시간"].dt.hour == 0, "하차 시간"] += pd.DateOffset(days=1)
```

```python
go_sort_data['이동 시간'] = go_sort_data['하차 시간'] - go_sort_data['탑승 시간']
go_sort_data
```
