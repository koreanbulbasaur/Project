```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
import requests
```

```python
def web_search(soup, sw):
    df = pd.DataFrame()
    movie_title=[]
    movie_point=[]

    for line in soup.findAll('div', class_='title_box'):
        title = line.find('strong', class_='name') 
        if title :
            movie_title.append(title.get_text().strip('\n'))
        
        point = line.find('span', class_='sub_text')
        if point :
            movie_point.append(point.get_text())
    df['Title'] = movie_title
    if sw == 1:
        df['Customer'] = movie_point
    else:
        df['Point'] = movie_point
    return df
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch'
url += '&query=현재상영영화순위&ie=utf8&sm=tab_she&qdt=0'

driver = webdriver.Chrome()
time.sleep(1)
driver.get(url)
xpath ='//*[@id="main_pack"]/div[2]/div[2]/div/div/div[1]/div/div/ul/li[1]'
items = driver.find_element(By.XPATH, xpath)
items.click()
time.sleep(1)
soup = BeautifulSoup(driver.page_source)
df1 = web_search(soup, 1)

xpath ='//*[@id="main_pack"]/div[2]/div[2]/div/div/div[1]/div/div/ul/li[2]'
items = driver.find_element(By.XPATH, xpath)
items.click()
time.sleep(1)
df2 = web_search(soup, 2)

driver.quit()
```

```python
merge_df = pd.merge(df1, df2, on='Title')
df = merge_df[merge_df['Customer'] != merge_df['Point']].drop_duplicates('Title')
df
```
